```python
def load_england_ae_file(file_path, period_date):
    """
    Load and clean one NHS England monthly A&E file.
    Handles both legacy .xls and modern .xlsx files (regardless of extension).
    """
    
    # Detect the actual file format from the first bytes
    with open(file_path, "rb") as f:
        first_bytes = f.read(8)
    
    if first_bytes[:2] == b"PK":
        # XLSX (zip-based)
        engine = "openpyxl"
    elif first_bytes[:4] == b"\xd0\xcf\x11\xe0":
        # Legacy XLS
        engine = "xlrd"
    else:
        raise ValueError(f"Unknown file format in {file_path}")
    
    # Safety check: does the expected sheet exist?
    xl = pd.ExcelFile(file_path, engine=engine)
    if "Provider Level Data" not in xl.sheet_names:
        raise ValueError(
            f"'Provider Level Data' sheet not found in {file_path}. "
            f"Sheets found: {xl.sheet_names}"
        )
    
    # Read the file
    df = pd.read_excel(
        file_path,
        sheet_name="Provider Level Data",
        engine=engine,
        skiprows=15
    )
    
    # Safety check 2: do the expected columns exist?
    required_cols = [
        "Code",
        "Region",
        "Name",
        "Type 1 Departments - Major A&E",
        "Type 1 Departments - Major A&E.2",
        "Percentage in 4 hours or less (type 1)",
        "Number of patients spending >12 hours from decision to admit to admission",
        "Emergency Admissions via Type 1 A&E",
        "Total Emergency Admissions"
    ]
    missing = [c for c in required_cols if c not in df.columns]
    if missing:
        raise ValueError(
            f"Missing expected columns in {file_path}: {missing}"
        )
    
    # Step 2 — Keep & rename
    df = df[required_cols].rename(columns={
        "Code": "trust_code",
        "Region": "trust_region",
        "Name": "trust_name",
        "Type 1 Departments - Major A&E": "attendances_type1",
        "Type 1 Departments - Major A&E.2": "breaches_4hr_type1",
        "Percentage in 4 hours or less (type 1)": "pct_4hr_type1",
        "Number of patients spending >12 hours from decision to admit to admission": "waits_12hr",
        "Emergency Admissions via Type 1 A&E": "admissions_type1",
        "Total Emergency Admissions": "total_admissions"
    })
    
    # Step 3 — Filter junk rows
    df = df[df["trust_code"].notna() & (df["trust_code"] != "-")]
    
    # Step 4 — Convert numeric columns
    numeric_cols = [
        "attendances_type1", "breaches_4hr_type1", "pct_4hr_type1",
        "admissions_type1", "total_admissions", "waits_12hr"
    ]
    for col in numeric_cols:
        df[col] = pd.to_numeric(df[col], errors="coerce")
    
    # Step 5 — Filter non-Type-1 trusts
    df = df[df["attendances_type1"] > 0]
    
    # Step 6 — Add period
    df["period"] = pd.to_datetime(period_date)
    
    # Safety check 3: did we get a sensible number of trusts?
    if len(df) < 50:
        print(
            f"⚠ Warning: only {len(df)} trusts in {file_path} — "
            f"may indicate a structure issue"
        )
    
    return df
```

```python
from pathlib import Path

era4_folder = Path("data/raw/england/era4_apr2024_may2026")

# List all .xls files in this folder
era4_files = list(era4_folder.glob("*.xls"))
print(f"Found {len(era4_files)} files")
for f in era4_files[:5]:
    print(f" - {f.name}")
```

```python
import re
from datetime import datetime

def extract_period_from_filename(filename):
    """
    Pull the reporting period out of an NHS filename.
    Handles both 'April-2024-AE-...' and 'April-26-AE-...' patterns.
    Returns a date string like '2024-04-01'.
    """
    months = {
        "january":1, "february":2, "march":3, "april":4, "may":5, "june":6,
        "july":7, "august":8, "september":9, "october":10, "november":11, "december":12
    }
    
    name = filename.lower()
    
    # Find month name
    month_num = None
    for m_name, m_num in months.items():
        if m_name in name:
            month_num = m_num
            break
    
    if month_num is None:
        raise ValueError(f"Could not find month name in {filename}")
    
    # Find year — try 4-digit first, then 2-digit
    year_match = re.search(r"(20\d{2})", name)
    if year_match:
        year = int(year_match.group(1))
    else:
        # Try 2-digit year like '-26-' or '-24-'
        year_match = re.search(r"-(\d{2})-", name)
        if year_match:
            year = 2000 + int(year_match.group(1))
        else:
            raise ValueError(f"Could not find year in {filename}")
    
    return f"{year}-{month_num:02d}-01"
```

```python
for f in era4_files[:5]:
    period = extract_period_from_filename(f.name)
    print(f"{f.name} → {period}")
```

```python
all_dataframes = []
failed_files = []

for file_path in era4_files:
    try:
        period = extract_period_from_filename(file_path.name)
        df_one = load_england_ae_file(str(file_path), period)
        all_dataframes.append(df_one)
        print(f"✓ {file_path.name} → {len(df_one)} trusts")
    except Exception as e:
        failed_files.append((file_path.name, str(e)))
        print(f"✗ {file_path.name} FAILED: {e}")

# Combine all successfully read files into one dataframe
era4_combined = pd.concat(all_dataframes, ignore_index=True)

print(f"\nTotal rows: {len(era4_combined)}")
print(f"Successful files: {len(all_dataframes)}")
print(f"Failed files: {len(failed_files)}")
```

```python
# All 4 era folders
era_folders = [
    Path("data/raw/england/era1_apr2019_mar2020"),
    Path("data/raw/england/era2_apr2020_mar2022"),
    Path("data/raw/england/era3_apr2022_mar2024"),
    Path("data/raw/england/era4_apr2024_may2026"),
]

all_dataframes = []
failed_files = []

for folder in era_folders:
    print(f"\n--- Processing {folder.name} ---")
    files = list(folder.glob("*.xls")) + list(folder.glob("*.xlsx"))
    
    for file_path in files:
        try:
            period = extract_period_from_filename(file_path.name)
            df_one = load_england_ae_file(str(file_path), period)
            all_dataframes.append(df_one)
            print(f"✓ {file_path.name} → {len(df_one)} trusts")
        except Exception as e:
            failed_files.append((file_path.name, str(type(e).__name__), str(e)))
            print(f"✗ {file_path.name} FAILED: {type(e).__name__}: {e}")

# Combine
england_all = pd.concat(all_dataframes, ignore_index=True)

print(f"\n=== SUMMARY ===")
print(f"Total rows: {len(england_all)}")
print(f"Successful files: {len(all_dataframes)}")
print(f"Failed files: {len(failed_files)}")
print(f"Date range: {england_all['period'].min()} to {england_all['period'].max()}")
print(f"Unique periods: {england_all['period'].nunique()}")
```

```python
for folder in era_folders:
    count = len(list(folder.glob("*.xls")) + list(folder.glob("*.xlsx")))
    expected = {"era1_apr2019_mar2020": 12, "era2_apr2020_mar2022": 24, 
                "era3_apr2022_mar2024": 24, "era4_apr2024_may2026": 26}
    expected_count = expected.get(folder.name, "?")
    status = "✓" if count == expected_count else "✗"
    print(f"{status} {folder.name}: {count} files (expected {expected_count})")
```
